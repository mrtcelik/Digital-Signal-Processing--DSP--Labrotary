```python
import numpy as np
import matplotlib.pyplot as plt 
```

## a) 

```python
# x1n

Fs=1000 
Ts=1/Fs 
# 200pi*N/1000 = 2pi*k ==> N = 10k, k = 1 için N = 10 
N=10 # örnek sayısı
n=np.arange(0.,2*N) 
x1n=7*np.cos(200*np.pi*n*Ts) 
plt.stem(n,x1n) 
plt.title("Örneklenen İşaret", ) 
plt.xlabel("n")
plt.ylabel("x1[n]") 
plt.show() 
```

```python
# x2n

Fs=1000 
Ts=1/Fs 
N=10 
n=np.arange(0.,2*N) 
x2n=3*np.cos(200*np.pi*n*Ts) + 4*np.cos(1800*np.pi*n*Ts)  
plt.stem(n,x2n)  
plt.title("Örneklenen İşaret", ) 
plt.xlabel("n")
plt.ylabel("x2[n]")
plt.show() 
```

## b) 

```python
import numpy as np
import matplotlib.pyplot as plt 

N=10 
w=np.linspace(-np.pi,np.pi-(2*np.pi/N),N) 
x_w=np.fft.fft(x1n,N)/N 
x_w=np.fft.fftshift(x_w) 

plt.figure() 
plt.stem(w,np.abs(x_w)) 
plt.title("X1(w)'nın Genlik Grafiği") 
plt.xlabel("w") 
plt.show() 
plt.figure() 
plt.stem(w,np.angle(x_w)) 
plt.title("X1(w)'nın Faz Grafiği") 
plt.xlabel("w") 
plt.show() 
```

```python
N=10 # örnek sayısı
w=np.linspace(-np.pi,np.pi-(2*np.pi/N),N) 
x_w=np.fft.fft(x2n,N)/N # x[n] işaretinin fourier transformunun 
x_w=np.fft.fftshift(x_w) 

plt.figure() 
plt.stem(w,np.abs(x_w)) 
plt.title("X2(w)'nın Genlik Grafiği") 
plt.xlabel("w") 
plt.show() 
plt.figure() 
plt.stem(w,np.angle(x_w)) 
plt.title("X2(w)'nın Faz Grafiği") 
plt.xlabel("w") 
plt.show() 
```

```python
# 𝑥[𝑛] = [1.25 , 3.5 , 4 , 6.7 , −1.4 , −3.5 , −10.5 , 8.5 , 3.8]


import numpy as np
import matplotlib.pyplot as plt 

def quantize(x, S):
    X = x.reshape((-1,1))
    S = S.reshape((1,-1))
    dists = abs(X-S)
    nearestIndex = dists.argmin(axis=1)
    quantized = S.flat[nearestIndex]
    return quantized.reshape(x.shape)

𝑥=np.array([1.25, 3.5, 4, 6.7, -1.4, -3.5, -10.5, 8.5, 3.8])
n = np.linspace(0,8,9)

k_bit= 3 
Max_range= np.amax(x)-np.amin(x) 
q=pow(2,k_bit) 
delta = Max_range/q 
S_midrise = -np.amax(x) + delta/2 + np.arange(q)*delta 
S_midtread = -np.amax(x) + np.arange(q)*delta 

y_midtrise = quantize(x, S_midrise) 
y_midtread = quantize(x, S_midtread) 

plt.figure()
fig, ax = plt.subplots()
ax.stem(n,x) 
ax.bar(n,y_midtrise,color='orange',edgecolor = "darkorange") 
plt.xlabel("örnek") 
plt.title('midrise')
fig, ax = plt.subplots()
ax.stem(n,x) 
ax.bar(n,y_midtread,color='orange',edgecolor = "darkorange") 
plt.xlabel("örnek") 
plt.title('midtread ')
plt.show() 
```
