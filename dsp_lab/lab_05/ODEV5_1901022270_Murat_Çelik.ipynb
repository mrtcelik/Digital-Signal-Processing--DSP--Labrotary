```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# x1(t) = 7cos(200*pi*t)

Fs=2200 # örnekleme frekansını
Ts=1/Fs # örnekleme periyodu

# 200pi*N/2200 = 2pi*k ==> N = 22k ==>k = 1 --> N = 22 
N=22 #Örnek Sayısı

n=np.arange(0.,2*N) # 2 Periyot

x1n=7*np.cos(200*np.pi*n*Ts) # t->n*Ts
plt.stem(n,x1n)
plt.title("Örneklenen İşaret", ) 
plt.xlabel("n")
plt.ylabel("x1[n]")
plt.show()
```

```python
# x2(t) = 3cos(200𝜋𝑡) + 4cos(1800𝜋𝑡)

Fs=2200 # örnekleme frekansı
Ts=1/Fs # örnekleme periyodu
# 200pi*N/2200 = 2pi*k ==> N = 22k, k = 1 için N = 22 
# 1800pi*N/2200 = 2pi*k ==> N = 22k/9, k = 9 için N = 22 
N=22 # örnek sayısı

n=np.arange(0.,2*N) # 2 Periyot
x2n=3*np.cos(200*np.pi*n*Ts) + 4*np.cos(1800*np.pi*n*Ts)  #örneklenmiş x[n] işareti
plt.stem(n,x2n) # x[n] işaretinin grafiğinin çizdirilmesi
plt.title("Örneklenen İşaret", ) 
plt.xlabel("n")
plt.ylabel("x2[n]")
plt.show()
```

## $ x_1(t)$ sinyalinin periyodu $T = \frac{1}{100}$ . $x_2(t)$ sinyalinin periyodu iki periyodik sinyalden oluşur ve periyotlarının ekoku yine $ T_2 =\frac{1}{100}$ 'dir. Böylece örnek sayıları eşittir. Bunun dışında farklı iki sinyal örneklendiğinden dolayı farklı sinyaller elde edilmiştir.

```python
# x1 İşareti için Fourier Transformu
N=22 # örnek sayısı
w=np.linspace(-np.pi,np.pi-(2*np.pi/N),N)
x_w=np.fft.fft(x1n,N)/N 
x_w=np.fft.fftshift(x_w)
plt.figure() 
plt.stem(w,np.abs(x_w)) 
plt.title("X1(w)'nın Genlik Grafiği") 
plt.xlabel("w") 
plt.show()  
plt.figure() 
plt.stem(w,np.angle(x_w)) 
plt.title("X1(w)'nın Faz Grafiği") 
plt.xlabel("w") 
plt.show() 
```

```python
# x2 İşareti için Fourier Transformu
N=22 
w=np.linspace(-np.pi,np.pi-(2*np.pi/N),N) 
x_w=np.fft.fft(x2n,N)/N
x_w=np.fft.fftshift(x_w) 
plt.figure() 
plt.stem(w,np.abs(x_w)) 
plt.title("X2(w)'nın Genlik Grafiği")
plt.xlabel("w") 
plt.show() 
plt.figure() 
plt.stem(w,np.angle(x_w))
plt.title("X2(w)'nın Faz Grafiği") 
plt.xlabel("w") 
plt.show() 
```

## $x_1(t) \Rightarrow x_1[n] = 7cos(\frac{\pi n}{11}) $ . Bu örnek sonucundan görüldüğü üzere sinyalin fourier transformu genliğinde $ w1 = \frac{\pi}{11}, w2 = \frac{- \pi}{11}$ olmak üzere iki noktada genliğinin yarısı (7/2) kadar değer alması beklenir

## $x_2(t) \Rightarrow x_2[n] = 3cos(\frac{\pi n}{11}) + 4sin(\frac{9 \pi n}{11}) $ .Bu örnek sonucundan görüldüğü üzere sinyalin fourier transformu genliğinde $ w1 = \frac{\pi}{11}, w2 = \frac{- \pi}{11}$ olmak üzere iki noktada genliğinin yarısı (3/2) kadar ve $ w3 = \frac{9 \pi}{11}, w4 = \frac{-9 \pi}{11}$ olmak üzere iki noktada genliğinin yarısı (4/2) kadar değer alması beklenir
