```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# x1(t) = 7cos(200*pi*t)

Fs=2200 # Ã¶rnekleme frekansÄ±nÄ±
Ts=1/Fs # Ã¶rnekleme periyodu

# 200pi*N/2200 = 2pi*k ==> N = 22k ==>k = 1 --> N = 22 
N=22 #Ã–rnek SayÄ±sÄ±

n=np.arange(0.,2*N) # 2 Periyot

x1n=7*np.cos(200*np.pi*n*Ts) # t->n*Ts
plt.stem(n,x1n)
plt.title("Ã–rneklenen Ä°ÅŸaret", ) 
plt.xlabel("n")
plt.ylabel("x1[n]")
plt.show()
```

```python
# x2(t) = 3cos(200ðœ‹ð‘¡) + 4cos(1800ðœ‹ð‘¡)

Fs=2200 # Ã¶rnekleme frekansÄ±
Ts=1/Fs # Ã¶rnekleme periyodu
# 200pi*N/2200 = 2pi*k ==> N = 22k, k = 1 iÃ§in N = 22 
# 1800pi*N/2200 = 2pi*k ==> N = 22k/9, k = 9 iÃ§in N = 22 
N=22 # Ã¶rnek sayÄ±sÄ±

n=np.arange(0.,2*N) # 2 Periyot
x2n=3*np.cos(200*np.pi*n*Ts) + 4*np.cos(1800*np.pi*n*Ts)  #Ã¶rneklenmiÅŸ x[n] iÅŸareti
plt.stem(n,x2n) # x[n] iÅŸaretinin grafiÄŸinin Ã§izdirilmesi
plt.title("Ã–rneklenen Ä°ÅŸaret", ) 
plt.xlabel("n")
plt.ylabel("x2[n]")
plt.show()
```

## $ x_1(t)$ sinyalinin periyodu $T = \frac{1}{100}$ . $x_2(t)$ sinyalinin periyodu iki periyodik sinyalden oluÅŸur ve periyotlarÄ±nÄ±n ekoku yine $ T_2 =\frac{1}{100}$ 'dir.
```python
# x1 Ä°ÅŸareti iÃ§in Fourier Transformu
N=22 # Ã¶rnek sayÄ±sÄ±
w=np.linspace(-np.pi,np.pi-(2*np.pi/N),N)
x_w=np.fft.fft(x1n,N)/N 
x_w=np.fft.fftshift(x_w)
plt.figure() 
plt.stem(w,np.abs(x_w)) 
plt.title("X1(w)'nÄ±n Genlik GrafiÄŸi") 
plt.xlabel("w") 
plt.show()  
plt.figure() 
plt.stem(w,np.angle(x_w)) 
plt.title("X1(w)'nÄ±n Faz GrafiÄŸi") 
plt.xlabel("w") 
plt.show() 
```

```python
# x2 Ä°ÅŸareti iÃ§in Fourier Transformu
N=22 
w=np.linspace(-np.pi,np.pi-(2*np.pi/N),N) 
x_w=np.fft.fft(x2n,N)/N
x_w=np.fft.fftshift(x_w) 
plt.figure() 
plt.stem(w,np.abs(x_w)) 
plt.title("X2(w)'nÄ±n Genlik GrafiÄŸi")
plt.xlabel("w") 
plt.show() 
plt.figure() 
plt.stem(w,np.angle(x_w))
plt.title("X2(w)'nÄ±n Faz GrafiÄŸi") 
plt.xlabel("w") 
plt.show() 
```
