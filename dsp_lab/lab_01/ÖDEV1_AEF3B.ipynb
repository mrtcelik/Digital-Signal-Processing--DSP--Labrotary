# ÖN HAZIRLIKLA İLGİLİ KODLAR

```python
#Bu Kod Bloğunda Ön Çalışmaya Ait Kodlar Bulunmaktadır
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
#2.1 Vektör dizisi oluşturmak

x=np.arange(0,10,2)
print(x)
x=np.linspace(0,10,5)
print(x)

#2.2 Ayrık-zamanlı işaret oluşturmak

n=np.arange(0,10)
x=np.cos(n*np.pi/3+np.pi/2)
print(x)

#2.3 Sürekli zamanlı işaret oluşturmak

t=np.linspace(0,2/50,1000)
x=np.cos(100*np.pi*t-np.pi/5)

#2.4 Grafik çizdirmek

n=np.arange(0,10)
x=np.cos(n*np.pi/3+np.pi/2)
print(x)
plt.stem(n,x)
plt.show()

plt.figure()
t=np.linspace(0,2/50,1000)
x=np.cos(100*np.pi*t-np.pi/5)
plt.plot(t,x)
plt.xlabel('zaman')
plt.ylabel('x(t)')
plt.show()

#2.5 Tek figürde birden fazla işaret çizdirmek

plt.figure()
t=np.linspace(0,2/50,1000)
x1=np.cos(100*np.pi*t-np.pi/5)
x2=2*np.sin(200*np.pi*t)
plt.plot(t,x1,t,x2)
plt.legend(('x1(t)','x2(t)'))
plt.show()

#2.6 İşaretleri ötelemek

plt.figure()
n=np.array([0,1,2,3,4])
x3=np.array([1,0,2,-1,0])
plt.xlabel('n')
plt.ylabel('x[n]')
plt.stem(n,x3)
plt.show()

plt.figure()
n=np.array([0,1,2,3,4])
n2=n+2
x3=np.array([1,0,2,-1,0])
plt.title('x[n-2]')
plt.stem(n2,x3)
plt.show()

n=np.array([0,1,2,3,4,5])
x=np.array([1,0,2,-1,0,0])
x_shifted=np.append(np.array([0,0]),x[:-2])
fig = plt.figure()#yeni figure
ax1=fig.add_subplot(211)
ax2=fig.add_subplot(212)
ax1.stem(n,x)
ax1.title.set_text('x[n]')
ax2.stem(n,x_shifted)
ax2.title.set_text('x[n-2]')

#2.7 Kompleks sayılar

x1=1+1j
x2=np.complex128(1+1j)
print(x1)
print(x2)
x=np.array([1+2j,2-6j],dtype=complex)
x=3+4j
print(np.abs(x))
x=np.array([1+2j,3-4j],dtype=complex)
print(np.abs(x))
x1=1+1j
print(np.angle(x1))
r=5
theta=np.pi/2
x=r*np.exp(1j*theta)
print(x)
```

```python
import numpy as np

a = 3 - 4j
b = 1 + 2j
D = np.divide(a, b)

print("Gerçek kısım:", D.real)#Real Kısım
print("Sanal kısım:", D.imag)#İmajiner Kısım
print("Genlik:", np.abs(D))#Genlik

faz_radyan = np.angle(D)#Radyan
faz_derece = np.rad2deg(faz_radyan)#Radyon to Degree

print("Faz (radyan):", faz_radyan)
print("Faz (derece):", faz_derece)

```

```python
import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib notebook
"""
Bu işaretlerin periyodunu bulmak için w0N = 2pik eşitliğini kullanırız. 
2sin(2pi/5n-pi/5) işaretinin frekansı 2pi/5 tir. 
Yerine koyarsak N=5 sample olarak buluruz. 
sin(3pin/5-2pi/5) işaretinin frekansı ise 3*pi/5 tir. 
Yerine koyarsak N=10 sample olarak bulunur. 
Bu iki işaretin periyotlarının OKEK ini alırsak x[n] işaretinin periyodunu buluruz. 
5 ve 10 un OKEK i 10 dur yani x[n] işaretinin periyodu 10 sample dır.
"""
plt.figure()
n=np.arange(0,20,1)#2 tam periyot için aralık
y=2*np.sin(2*np.pi*n/5-np.pi/5)+np.sin(3*np.pi*n/5-2*np.pi/5) #İşaret
plt.stem(n,y)
plt.xlabel("n")
plt.ylabel("x[n]")
plt.title("X[n] İşareti")
plt.axvspan(0,9,color=("red"),alpha=0.5)#Periyodu
plt.show()
```

# SORU-3

```python
import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib notebook

n=np.arange(0,16)#İstenen Aralık
y=(np.exp(1j*np.pi/5))

#Aşağıda kullanılabilir bir array olarak kullanmak için gereklidir
yreel = np.array(np.zeros(len(n))) #Reel kısım için.
yimag = np.array(np.zeros(len(n))) #İmajiner kısım için.
genlik = np.array(np.zeros(len(n))) #Genlik için
faz = np.array(np.zeros(len(n))) #Faz için
for i in range(0,16,1): #Döngü ile Atama
    yreel[i] = np.real(y) #Real Kısım Atama
    yimag[i] = np.imag(y) #Sanal Kısım Atama
    genlik[i] = np.abs(y) #Genlik Atama
    faz[i] = np.angle(y)  #Faz Atama
plt.figure()
plt.stem(n,yreel)
plt.title("x[n] = e^(j*pi/5) işaretinin gerçek kısmının grafiği")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
plt.figure()
plt.stem(n,yimag)
plt.title("x[n] = e^(j*pi/5) işaretinin sanal kısmının grafiği")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
plt.figure()
plt.stem(n,genlik)
plt.title("x[n] = e^(j*pi/5) işaretinin genlik grafiği")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
plt.figure()
plt.stem(n,faz)
plt.title("x[n] = e^(j*pi/5) işaretinin faz grafiği")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
```
