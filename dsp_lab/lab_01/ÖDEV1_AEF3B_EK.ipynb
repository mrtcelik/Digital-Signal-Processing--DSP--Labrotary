# Soru-1

```python
import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib notebook
a=3-4j
b=1+2j
c=a+b

real=np.real(c)#reel
imag=np.imag(c)#sanal
genlik=abs(c)#genlik
faz=np.angle(c)#fazı(radyan cinsinden)
fazderece=np.angle(c,True)#fazı(derece cinsinden)
print("reel kısmı: {}".format(real))
print("sanal kısmı: {}".format(imag))
print("genlik: {}".format(genlik))
print("fazı(radyan cinsinden): {} radyan".format(faz))
print("fazı(derece cinsinden): {} derece".format(fazderece))
```

# Soru-2

```python
import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib notebook
"""
2sin(2pi.n/5-pi/5) işaretinin frekansı 2pi/5 tir. 
Yerine koyarsak N=5 sample olarak buluruz. 
sin(3pi.n/5-2pi/5) işaretinin frekansı ise 3*pi/5 tir. 
Yerine koyarsak N=10 sample olarak bulunur.
"""
n=np.arange(0,10)#İlgili Aralık
y1=2*np.sin(2*np.pi*n/5-np.pi/5) #İşaret
plt.stem(n,y1)
plt.grid(True)
plt.title("X[n]=2*sin(2*pi/5*n-pi/5) işaretinin grafiği")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.axvspan(0,4,color=("red"),alpha=0.25)#Periyod
```

```python
#İkinci Kısım

import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib notebook
plt.figure()
n=np.arange(0,20)#İlgili Aralık.
y2=np.sin(3*np.pi*n/5-2*np.pi/5)
plt.stem(n,y2)
plt.grid(True)
plt.title("X[n]=sin(3*pi/5*n-pi/5) işaretinin grafiği")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.axvspan(0,9,color=("yellow"),alpha=0.25)#Periyo
```

```python
#BİRİNCİ KISIM

import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib notebook
n=np.arange(0,16)#Aralık
y=np.array(np.zeros(len(n)))
for i in range(0,16,1):
    y[i] = (np.exp(np.pi/5))

plt.figure()
plt.stem(n,y)
plt.title("x1[n] = e^(pi/5) işaretinin grafiği")
plt.xlabel("n")
plt.ylabel("x1[n]")
plt.grid(True)
```

```python
#İKİNCİ KISIM

import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib notebook
n=np.arange(0,16)
y=np.array(np.zeros(len(n)))#Kontrol Dizisi
yimag=np.array(np.zeros(len(n)))#Sanal Kısım Dizisi
genlik=np.array(np.zeros(len(n)))#Genlik 
faz=np.array(np.zeros(len(n)))#Faz
for i in range(0,16,1):
    y[i] = (np.exp(1j*np.pi/5*n[i])).real #İlgili Atama
    yimag[i] = (np.exp(1j*np.pi/5*n[i])).imag 
    genlik[i] = np.abs(np.exp(1j*np.pi/5*n[i]))
    faz[i] = np.angle(np.exp(1j*np.pi*n[i]/5))
plt.figure()
plt.stem(n,y)
plt.title("x3[n] = e^(j*pi/5*n) işaretinin gerçek kısmının grafiği")
plt.xlabel("n")
plt.ylabel("x3[n]")
plt.grid(True)
plt.axvspan(0,9,color=("red"),alpha=0.5)
plt.figure()
plt.stem(n,yimag)
plt.title("x3[n] = e^(j*pi/5*n) işaretinin sanal kısmının grafiği")
plt.xlabel("n")
plt.ylabel("x3[n]")
plt.grid(True)   
plt.axvspan(0,9,color=("red"),alpha=0.25)

plt.figure()
plt.stem(n,genlik)
plt.title("x3[n] = e^(j*pi/5*n) işaretinin genlik grafiği")
plt.xlabel("n")
plt.ylabel("x3[n]")
plt.grid(True)
plt.figure()
plt.stem(n,faz)
plt.title("x3[n] = e^(j*pi/5*n) işaretinin faz grafiği")
plt.xlabel("n")
plt.ylabel("x3[n]")
plt.grid(True)
faz
```

```python
#ÜCÜNCÜ KISIM

import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib notebook
n=np.arange(0,16)
y=np.cos(np.pi/5)+1j*np.sin(np.pi/5)
yreal=np.array(np.zeros(len(n)))#Kontrol Dizisi
yimag=np.array(np.zeros(len(n)))
genlik=np.array(np.zeros(len(n)))
faz=np.array(np.zeros(len(n)))
for i in range(0,16,1):
    yreal[i]=np.real(y)
    yimag[i]=np.imag(y)
    genlik[i]=np.abs(y)
    faz[i]=np.angle(y)
plt.figure()
plt.stem(n,yreal)
plt.title("x4[n] = cos(pi/5)+1j*sin(pi/5) işaretinin gerçek kısmının grafiği")
plt.xlabel("n")
plt.ylabel("x4[n]")
plt.grid(True)
plt.figure()
plt.stem(n,yimag)
plt.title("x4[n] = cos(pi/5)+1j*sin(pi/5) işaretinin sanal kısmının grafiği")
plt.xlabel("n")
plt.ylabel("x4[n]")
plt.grid(True)    

plt.figure()
plt.stem(n,genlik)
plt.title("x4[n] = cos(pi*n/5)+1j*sin(pi*n/5) işaretinin genlik grafiği")
plt.xlabel("n")
plt.ylabel("x4[n]")
plt.grid(True)

plt.figure()
plt.stem(n,faz)
plt.title("x4[n] = cos(pi*n/5)+1j*sin(pi*n/5) işaretinin faz grafiği")
plt.xlabel("n")
plt.ylabel("x4[n]")
plt.grid(True)
```
