###  Girişine sinüsodial bir işaret uygulanan bir sistemi konvolüsyon işlemi yapmadan özfonskiyonlardan yararlanarak kolaylıkla elde edebiliriz. Çıkış işaretinde genlik |H(e^jw)|'nin  genliği kadar değişir ve H(e^jw)'nin faz açısı işarete faz olarak işarete eklenir

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
from scipy.fftpack import fft , ifft

n=np.arange(0,11)
h=np.zeros(len(n))
y=np.zeros(len(n))

h[0]=1
h[1]=-1



dft_h=fft(h)
dft_amp=np.abs(dft_h)
dft_phase=np.angle(dft_h)
print("***Genlik Değerleri***")
print(dft_amp)
print("***Faz Değerleri***")
print(dft_phase)

for i in n:
    y[i]=(dft_amp[i])*(np.cos(((np.pi/3)*i)+(dft_phase[i])))

plt.figure()
plt.plot(n,y)
plt.title('y[n]')
plt.xlabel('n')

```
